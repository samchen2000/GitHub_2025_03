## 這裡是一個建立`正弦波`圖表的範例
## 顯示不同的格式的**文字**說明

```python
import numpy as np
import matplotlib.pyplot as plt

# 建立 x 軸的數據點，從 0 到 2π，間隔為 0.1
x = np.arange(0, 2*np.pi, 0.1)

# 使用 sin() 函數計算 y 軸的數據點
y = np.sin(x)

# 建立圖表
plt.plot(x, y)

# 設定圖表標題和軸標籤
plt.title('sin polt')
plt.xlabel('x line')
plt.ylabel('y line')

# 顯示圖表
plt.show()
```

這裡是一個建立余弦波圖表的範例

```python
import numpy as np
import matplotlib.pyplot as plt

# 建立 x 軸的數據點，從 0 到 2π，間隔為 0.1
x = np.arange(0, 2*np.pi, 0.1)

# 使用 cos() 函數計算 y 軸的數據點
y = np.cos(x)

# 建立圖表
plt.plot(x, y)

# 設定圖表標題和軸標籤
plt.title('Cos Polt')
plt.xlabel('x line_1')
plt.ylabel('y line')

# 顯示圖表
plt.show()
```

同時顯示sin() Cos()的正弦波與餘弦波的圖表

```python
import numpy as np
import matplotlib.pyplot as plt

# 生成數據
x = np.linspace(0, 2 * np.pi, 100)
y_sin = np.sin(x)
y_cos = np.cos(x)

# 繪製圖表
plt.figure(figsize=(10, 5))
plt.plot(x, y_sin, label='sin(x)')
plt.plot(x, y_cos, label='cos(x)')
plt.title('Sine and Cosine Waves')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()
```
