## 這裡是一個建立`正弦波`圖表的範例
## 顯示不同的格式的**文字**說明

```python
import numpy as np
import matplotlib.pyplot as plt

# 建立 x 軸的數據點，從 0 到 2π，間隔為 0.1
x = np.arange(0, 2*np.pi, 0.1)

# 使用 sin() 函數計算 y 軸的數據點
y = np.sin(x)

# 建立圖表
plt.plot(x, y)

# 設定圖表標題和軸標籤
plt.title('sin polt')
plt.xlabel('x line')
plt.ylabel('y line')

# 顯示圖表
plt.show()
```
